```python
def display_side_by_side(*args, titles=cycle([''])):
    html_str = ''
    for df, title in zip(args, chain(titles, cycle(['</br>']))):
        html_str += '<th style="text-align:center"><td style="vertical-align:top">'
        html_str += f'<h2 style="color:#2f6eab; text-align:center">{title}</h2>'
        html_str += df.to_html().replace('table', 'table style="display:inline"')
        html_str += '</td></th>'
    display_html(html_str, raw=True)
```

Функция вывода info (расширенный вариант с выводом уникальных значений и диапазоном значений)
    

```python
cmap = list(reversed(
    ['#220a06',
     '#220a06',
     '#854b42',
     '#aa6055',
     '#d0776a',
     '#f0a99e',
     '#fbc1b9',
     '#ffe8e5',
     '#f8f1f0',
     '#fdf7f6']))
def data_info(data, full=True):    

    info = pd.DataFrame(
        [
            data.columns,
            data.isna().sum(),
            [data[col].dtype for col in data.columns],
            [data[col].unique() if(data[col].dtype == object) 
             else (f'max={data[col].max()} mean={data[col].mean():,.2f} median={data[col].median():,.2f} min={data[col].min()} ')
             for col in data.columns]

        ]
    ).transpose()
    MisVal = f'MisVal  ({data.isna().sum().sum()} out of {len(data)* len(data.columns)})'
    Columns = f'Columns  shape:{data.shape}'
    info.columns = [Columns, MisVal, 'dtype', 'Val(unique/distr)']
    info.set_index(Columns, inplace=True)
    if full == False:
        info = info.loc[info[MisVal] != 0]
    return info.style.background_gradient(axis=0, cmap='Greens', low=0, high=0.2)
```

Сформирруем дополнительные синтетические признаки на основании данных датафрейма **contracts**
- **SinceBeginByCharge** количество месяцев, рассчитанное исходя из общего и помесячного расхода клиента
    - $SinceBeginByCharges = TotalCharges / MonthlyCharges$
- **SinceBeginDate** количество лет, прошедшее с момента открытия контракта 
    - $SinceBeginDate = год_{текущий} - год_{BeginDate}$
- **ChargesGropup** группы клиентов по общему объему расходования<br> 
    - $Группа 1$ - объем расходов < 25% квантиля<br>
    - $Группа 2$ - объем расходов < 75% квантиля<br>
    - $Группа 3$ - объем расходов >= 75% квантиля<br>  
    
</div>    

```python
# в ходе анализа выявлено, что признак TotalCharges содержит строковое значение " "
# заменим его на 0
contracts.loc[contracts['TotalCharges'] == " ", 'TotalCharges'] = 0

# преобразуем тип данных TotalCharges в float
contracts['TotalCharges'] = contracts['TotalCharges'].astype('float64')

# сформируем синтетический признак SinceBeginByCharges 
contracts['SinceBeginByCharges'] = round(contracts['TotalCharges'] / contracts['MonthlyCharges'])

# сформируем синтетический признак SinceBeginByCharges
contracts['SinceBeginDate'] = (2020 - contracts['BeginDate'].str[0:4].astype('int64')) * 12

# сформируем синтетический признак ChargesGropup
contracts.loc[contracts['TotalCharges'] >= contracts['TotalCharges'].quantile(0.75), 'ChargesGropup'] = 'Top'
contracts.loc[contracts['TotalCharges'] < contracts['TotalCharges'].quantile(0.75), 'ChargesGropup'] = 'Hi'
contracts.loc[contracts['TotalCharges'] < contracts['TotalCharges'].quantile(0.50), 'ChargesGropup'] = 'Middle'
contracts.loc[contracts['TotalCharges'] < contracts['TotalCharges'].quantile(0.25), 'ChargesGropup'] = 'Low'

contracts.loc[contracts['SinceBeginDate'] > contracts['SinceBeginByCharges'], 'MoneyTime'] = 'Additional'
contracts.loc[contracts['SinceBeginDate'] <= contracts['SinceBeginByCharges'], 'MoneyTime'] = 'Regular'


display(contracts.head())
print(contracts.info())
```

- при при отсутствии даты в столбце **EndDate** - клиент продолжает пользоваться услугами оператора  **Customer_left** = **0**

Также рассмотрим распределение численных показателей - признаков **MonthlyCharges** и **TotalCharges**. Дя чего визуализируем расрпеделение этих показателей с помощью гистограмм и гарафиков размаха. Построение графиков реализуем отдельной функцией **distr_graph**. На графиках выделим и обозначим границы зон повышенных рисков, области где отток клиентов достигал наибольших значений.

```python
def distr_graph(gr_data, distr_feature, hue_feature, step=10, risk_zone=[0.0, 0.0, 0.0]):
    sns.set(font_scale=1.33, style={"grid.color": ".7", "grid.linestyle": ":"})
    fig, ax = plt.subplots(1, 2,
                           figsize=(18, 10),
                           gridspec_kw={'width_ratios': [3, 1]})

    # Гистограмма распределения
    sns.histplot(
        data=gr_data,
        x=distr_feature,
        hue=hue_feature,
        palette=["#9a9c9d", "#ca2f1d"],
        kde=True,
        alpha=0.2,
        ax=ax[0])

    # Границы зоны рисков гистограммы
    for q in [risk_zone[0], risk_zone[1]]:
        ax[0].axvline(x=q, ls='--', color="#ca2f1d", alpha=0.3, linewidth=2)
    ax[0].axhline(y=risk_zone[2], ls='--', color="#ca2f1d", alpha=0.3, linewidth=2)

    # настройки осей и легенды  гистограммы
    ax[0].xaxis.set_major_locator(ticker.MultipleLocator(step))
    ax[0].yaxis.set_major_locator(ticker.MultipleLocator(100))
    ax[0].set_title(f'Распределение показателя {distr_feature}',
                    color='#985431',
                    fontsize=30)
    ax[0].set_ylabel('Count', fontsize=18, color="#985431")
    ax[0].set_xlabel(distr_feature, fontsize=18, color="#985431")
    ax[0].legend(labels=['1 - Ушедшие', '0 - Активные', 'Границы зоны повышенных рисков'],
                 loc='upper right',
                 fontsize=16)
    
    # График размаха
    sns.boxplot(
        data=gr_data,
        x=hue_feature,
        y=distr_feature,
        hue=hue_feature,
        palette=["#9a9c9d", "#ca2f1d"],
        boxprops=dict(alpha=.3),
        ax=ax[1]
    )
    
    # Границы зоны рисков график размаха   
    for q in [risk_zone[0], risk_zone[1]]:
        ax[1].axhline(
            y=q,
            ls='--',
            color="#ca2f1d",
            alpha=0.4,
            linewidth=2)
    
    # настройки осей и легенды график размаха    
    ax[1].yaxis.set_major_locator(ticker.MultipleLocator(step))
    ax[1].set_ylabel(f'{distr_feature}', fontsize=18, color="#985431")
    ax[1].set_xlabel(f'{hue_feature}', fontsize=18, color="#985431")
    ax[1].legend([], [], frameon=False)
    plt.subplots_adjust(hspace=0.2)
    plt.show()
```

Произведем два типа кодирования:
- **One-Hot Encoding** - прямое кодирование (для использования в линейной модели) с помощью **OneHotEncoder()** библиотеки **sklearn**
- **Ordinal Encoding** - порядковое кодирование (для моделей на основе дерева решений и бустингов) с помощью метода предобработки данных **OrdinalEncoder()** библиотеки **sklearn**
    

    
    
А также исключим из массивов **ohe_features_train, ohe_features_test** (для линейных моделей) список мультиуалинеарных показателей **drop_list**

```python
#кодирование Ordinal
ord_features_train = features_train.copy()
ord_features_test = features_test.copy()
ord_encoder = OrdinalEncoder()
ord_cat_features = [col for col in list(ord_features_train.columns) if ord_features_train[col].dtype == object]
ord_encoder.fit(ord_features_train[ord_cat_features])
ord_features_train[ord_cat_features] = ord_encoder.transform(ord_features_train[ord_cat_features])
ord_features_test[ord_cat_features] = ord_encoder.transform(ord_features_test[ord_cat_features])

#кодирование OHE
ohe_features_train = features_train.copy().drop(drop_list, axis=1)
ohe_features_test = features_test.copy().drop(drop_list, axis=1)
ohe_cat_features = [col for col in list(ohe_features_train.columns) if ohe_features_train[col].dtype == object]

transformer = make_column_transformer(
    (OneHotEncoder(), ohe_cat_features),
    remainder='passthrough')

ohe_features_train = pd.DataFrame(transformer.fit_transform(ohe_features_train), columns=transformer.get_feature_names())
ohe_features_test = pd.DataFrame(transformer.transform(ohe_features_test), columns=transformer.get_feature_names())

print(f'ord_features_train, ord_features_test, target_train, target_test')
print(ord_features_train.shape, ord_features_test.shape, target_train.shape, target_test.shape, '\n')
print(f'ohe_features_train, ohe_features_test, target_train, target_test')
print(ohe_features_train.shape, ohe_features_test.shape, target_train.shape, target_test.shape)
```

Для этого воспользуемся  вспомогательным классом **make_scorer** модуля **metrics** библиотеки **sklearn**. Для анализа качества будем использовать следующий набор метрик качества:

    - **accuracy**
    - **f1**    
    - **precision**
    - **recall**

Реализуем функцию обучения моделей и подбора гиперпараметров **model_tuning** на основе вспомогательного класса **RandomizedSearchCV** модуля **model_selection** библиотеки **sklearn**, на вход будет принимать следующие параметры:
- **default_model** - начальная модель
- **params** - перечень параметров с заданными диапазонами значений, которые будут передаваться для подбора RandomizedSearchCV 
- **train_pool** - обучающий пул данных(список **[массив признаков, целевой признак]**)
- **iterations** - количество итераций

**Функция будет осуществлять:**
- обучение модели, кроссвалидацию и подбор гиперпараметров
- сохранение и обновление(в случае улучшения основной метрики) обученной модели в переменной **global_project_models**
- формирование датафрейма **global_models_summary**, где будет сохраняться и обновлятся список всех моделей с лучшими метриками

**Обоснование:**
- Использование **RandomizedSearchCV** позволяет рассмотреть более широкий диапазон параметров в более короткие сроки
- Функция позволяет более вариативно фрагментировать временные интервалы подбора гиперпараметров модели. Наряду с параметром   **RandomizedSearchCV** **n_iter**, устанавливающим число проходов по диапазону параметров, мы получаем внешний счетчик итераций функции **iterations**, которым можем дополнительно фрагментировать процедуру, в целях более детализрованного мониторинга процесса подбора гиперпараметров
- формирование глобальных переменных для хранения моделей и статистики, позволяет быстро и удобно обращаться и к оптимально обученной модели и к ее показателям
    

```python
def model_tuning(default_model, params, train_pool, iterations = 1):   
    for i in range(iterations):
        #Зададим глобальные переменные для хранения моделей и статистики
        global global_project_models, global_models_summary

        # подбор гиперпараметров, кроссвалидация 
        model = default_model
        model_gs = RandomizedSearchCV(
            model,
            params,
            n_iter=3,
            n_jobs=-1,
            cv=cv,
            refit="roc_auc",
            scoring=scoring,
            random_state=random_state
        )
       
        model_gs.fit(train_pool[0], train_pool[1])

        # формирование имени переменной исходя из названия алгоритма модели
        model_name = str(locals()['model']).split('(')[0] + '_model'

        # сохраняем в переменную лучшую модель
        locals()[model_name] = model_gs.best_estimator_
        
        # Проверка существования глобальных переменных на случай перезаписи
        # Пробуем измерить длинну переменной
        try:
            len(global_project_models) 
        except:
            global_project_models = {}
        try:
            len(global_models_summary)
        except:
            global_models_summary = pd.DataFrame()

        # пробуем обновить модель и статистику если метрика улучшилась
        # в сравнении с существующей записью
        try:
            if global_models_summary.loc[model_name]['ROC-AUC'] < model_gs.best_score_:

                global_models_summary.drop([model_name], axis=0, inplace=True)
                
                global_models_summary =  pd.concat([global_models_summary, 
                pd.DataFrame(
                    [
                        [model_gs.best_score_],
                        [model_gs.cv_results_['mean_test_accuracy'][0]],
                        [model_gs.cv_results_['mean_test_f1'][0]],
                        [model_gs.cv_results_['mean_test_precision'][0]],
                        [model_gs.cv_results_['mean_test_recall'][0]],
                        
                    ],
                    index=['ROC-AUC', "Accuracy", "f1", "Precision", "Recall"],
                    columns = [model_name]
                ).transpose()], axis= 0)
                global_project_models[model_name] = locals()[model_name]

        # добавляем модель и статистику если запись о модели не создавалась      
        except:
            global_models_summary =  pd.concat([global_models_summary, 
                pd.DataFrame(
                    [
                        [model_gs.best_score_],
                        [model_gs.cv_results_['mean_test_accuracy'][0]],
                        [model_gs.cv_results_['mean_test_f1'][0]],
                        [model_gs.cv_results_['mean_test_precision'][0]],
                        [model_gs.cv_results_['mean_test_recall'][0]],
                        
                    ],
                    index=['ROC-AUC', "Accuracy", "f1", "Precision", "Recall"],
                    columns = [model_name]
                ).transpose()], axis= 0)
            global_project_models[model_name] = locals()[model_name]
#             print(model_gs.cv_results_)
        
        # Итоговый вывод статистики (в процессе)
        display(global_models_summary.sort_values(['ROC-AUC'], ascending=False).style.background_gradient(axis=0, cmap='bone'))
#         print(f"Сформирована модель {model_name} и обновлена запись в словаре global_project_models")
        print(f"Идет подбор гиперпараметров ... (выполнено итераций {i+1}/{iterations})")
        clear_output(wait=True)
        
    # Итоговый вывод статистики (по завершении)    
    display(global_models_summary.sort_values(['ROC-AUC'], ascending=False).style.background_gradient(axis=0, cmap='bone'))
    print(f"Подбор гиперпараметров завершен")
    print(f"Имя модели {model_name} (оптимально обученная модификация сохранена в global_project_models)")
#     print(f"Подбор гиперпараметров завершен")
```

Визуализируем **roc** кривую двух лучших моделей. График смоделированный по этапам кросвалидации и средним значением.

```python
mod_list = [global_project_models['GradientBoostingClassifier_model'], global_project_models['RandomForestClassifier_model']]
mod_name = ['GradientBoostingClassifier_model', 'RandomForestClassifier_model']
sns.set(font_scale=1.4, style="whitegrid")
fig, ax = plt.subplots(1, 2, figsize=(20, 10),  gridspec_kw={'width_ratios': [1, 1]})

for num in range(len(mod_list)):
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)
    X=np.array(ord_train_pool[0])
    y=np.array(ord_train_pool[1])

    for i, (train, test) in enumerate(cv.split(X, y)):
        mod_list[num].fit(X[train], y[train])

        viz = RocCurveDisplay.from_estimator(
            mod_list[num],
            X[test],
            y[test],
            name=f"ROC fold {i+1}",
            alpha=0.3,
            lw=1,
            ax=ax[num],
        )
        interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
        interp_tpr[0] = 0.0
        tprs.append(interp_tpr)
        aucs.append(viz.roc_auc)

    ax[num].plot([0, 1], [0, 1], linestyle="--", lw=2, color='#ca2f1d', label="Chance", alpha=0.5)

    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    ax[num].plot(
        mean_fpr,
        mean_tpr,
        color="b",
        label=f"Mean ROC-AUC = {round(mean_auc, 2)}",
        lw=2,
        alpha=0.8,
    )

    ax[num].set(
        xlim=[-0.05, 1.05],
        ylim=[-0.05, 1.05],
        title=f"{mod_name[num]} = {round(mean_auc, 2)}",
    )
    ax[num].legend(loc="lower right")
plt.show()
```
